2D FDM HEAT DIFFUSION

```python
# DIV = divergence \/.
# GRAD = gradient \/f(p) = [df/dx1 ... df/dxn]
# Laplace equation \/^2f = d2fdx2 + d2fdy2 + d2fdz2 = 0

#d2T/dx2 + d2Tdy2 = 0

#central difference numerical scheme
# (T(i+1)j - 2Tij + T(i-1)j) / dx2 + (Ti(j+1) - 2Tij + Ti(j-1) / dy2 = 0
# Tij = 1/4 * (T(i+1)j + T(i-1)j + Ti(j+1) + Ti(j-1))

#i -> x,  j -> y

```

```python
import numpy as np
import matplotlib.pyplot as plt


#===================================================================================================
# HEAT DIFFUSION CLASS
#===================================================================================================

class heaters:
    
    def __init__(self,lenHx,lenHy,locX,locY,bc_heater):
        self.lenHx = lenHx
        self.lenHy = lenHy
        self.locX = locX
        self.locY = locY
        self.bc_heater = bc_heater
    
class Plate:
    
    #class attribute
    number_of_heaters = 0
    
    #numHeaters must depend on add_heater
    
    def __init__(self, lenX,lenY,delta,maxIter):
        self.lenX = lenX
        self.lenY = lenY
        self.delta = delta
        self.maxIter = maxIter
        self.temp_array = np.empty((self.lenX,self.lenY))
        self.heater_array = []

    def add_heater(self, heaters):
        self.heater_array.append(heaters)
        Plate.add_heater_count()         
    
    def make_init_array(self, temp_init):
        self.temp_array.fill(temp_init)
                 
    def make_grid(self):
        X,Y = np.meshgrid(np.arange(0,self.lenX),np.arange(0,self.lenY))
        return X, Y
    
    def iterate(self):
        for iteration in range(0,self.maxIter):
            for i in range(1,self.lenX-1,self.delta):
                for j in range(1,self.lenY-1,self.delta):
                    self.temp_array[i,j] = 0.25*(self.temp_array[i+1,j]+self.temp_array[i-1,j]+self.temp_array[i,j+1]+self.temp_array[i,j-1])
        return self.temp_array
    
    @classmethod
    def add_heater_count(cls):
        cls.number_of_heaters += 1
        
    @classmethod
    def numHeater(cls):
        return cls.number_of_heaters
    
        
    #generate boundary conditions
    
    def add_boundaries(self):   
        
        for i in range(Plate.number_of_heaters):
            lenHx = self.heater_array[i].lenHx
            lenHy = self.heater_array[i].lenHy
            locX = self.heater_array[i].locX
            locY = self.heater_array[i].locY
            bc_heater = self.heater_array[i].bc_heater
            self.temp_array[locX:(locX+lenHx),locY:(locY+lenHy)] = bc_heater
            
    def plot_graphics(self):
        X, Y = self.make_grid()
        colourInterp = 50
        colourMap = plt.cm.gnuplot
        plt.title("2D Laplace thermal diffusion")
        plt.contourf(X,Y,self.temp_array,colourInterp,cmap = colourMap)
        plt.colorbar()
        plt.rcParams["figure.figsize"] = (10,8)
        plt.rcParams.update({'font.size': 10})
        plt.show()
            
            
#===================================================================================================
# MAIN
#===================================================================================================
if __name__ == "__main__":
    try:
        h1 = heaters(100,25,50,20,60)
        h2 = heaters(100,25,50,55,60)
        h3 = heaters(100,25,50,87,60)
        h4 = heaters(100,25,50,120,60)
        h5 = heaters(100,25,50,155,60)
        h6 = heaters(15,130,170,35,60)
        h7 = heaters(15,130,15,35,60)

        p1 = Plate(200,200,1,100)

        p1.add_heater(h1)
        p1.add_heater(h2)
        p1.add_heater(h3)
        p1.add_heater(h4)
        p1.add_heater(h5)
        p1.add_heater(h6)
        p1.add_heater(h7)


        p1.make_init_array(-60)
        p1.add_boundaries()
        p1.iterate()
        p1.plot_graphics()

    except:
        print("error occured")
```
